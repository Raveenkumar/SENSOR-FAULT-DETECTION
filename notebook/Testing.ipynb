```python
from box import ConfigBox
from pathlib import Path
import json
import os,sys
import pandas as pd
from src.logger import logger
from src.exception import SensorFaultException

def read_json(file_path:Path) -> ConfigBox:
    """read_json: used for reading json file 

    Args:
        file_path (Path): file_path

    Returns:
        ConfigBox | None: Provide ConfigBox (contains json data)
    """
    try:
        if os.path.exists(file_path):
            with open(file_path,mode='r') as file:
                json_data = json.load(file)
            logger.info(f"read json data : file_path: {file_path} : Status: Successful")    
        return ConfigBox(json_data)    
    
    except Exception as e:
       logger.error(msg=SensorFaultException(error_message=e,error_detail=sys))
       raise SensorFaultException(error_message=e,error_detail=sys)   
```

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import Border, Side, PatternFill, Font, Alignment
from datetime import datetime

# Function to apply styling to the Excel file
def style_excel(excel_filename):
    # Load the workbook and select the active worksheet
    wb = load_workbook(excel_filename)
    ws = wb.active

    # Define border style
    thin_border = Border(
        left=Side(style='thin'),
        right=Side(style='thin'),
        top=Side(style='thin'),
        bottom=Side(style='thin')
    )

    # Define header fill color and font style
    header_fill = PatternFill(start_color='FFFF00', end_color='FFFF00', fill_type='solid')  # Yellow color
    header_font = Font(bold=True, color='000000')  # Black color for text

    # Apply styles to the header row
    for cell in ws[1]:
        cell.fill = header_fill
        cell.font = header_font
        cell.border = thin_border
        cell.alignment = Alignment(horizontal='center', vertical='center')

    # Apply border to all data cells and adjust column widths
    for row in ws.iter_rows(min_row=2, min_col=1, max_col=ws.max_column, max_row=ws.max_row):
        for cell in row:
            cell.border = thin_border

    # Adjust column widths based on the maximum length of the data in each column
    for column in ws.columns:
        max_length = 0
        column_letter = column[0].column_letter  # Get the column letter (e.g., 'A', 'B')
        for cell in column:
            try:
                max_length = max(max_length, len(str(cell.value)))
            except:
                pass
        adjusted_width = max_length + 2  # Add some padding
        ws.column_dimensions[column_letter].width = adjusted_width

    # Save the styled workbook
    wb.save(excel_filename)
    print(f"Excel file '{excel_filename}' styled successfully with adjusted column widths.")

# Function to append logs to the existing Excel file with automatic slno
def append_log(filename, status, remark, batchfolder, excel_filename='log_file.xlsx'):
    try:
        # Load the existing data into a DataFrame if the file exists
        existing_df = pd.read_excel(excel_filename)
        
        # Define slno automatically based on the number of existing rows
        slno = len(existing_df) + 1
        
        # Get the current date for the DATE column
        current_date = datetime.now().strftime('%Y-%m-%d')
        
        # Create a DataFrame for the new log entry
        new_data = {
            'SLNO': [slno],
            'DATE': [current_date],
            'BATCHFOLDER': [batchfolder],
            'FILENAME': [filename],
            'STATUS': [status],
            'REMARK': [remark]
        }
        new_df = pd.DataFrame(new_data)
        
        # Concatenate the existing DataFrame with the new DataFrame
        updated_df = pd.concat([existing_df, new_df], ignore_index=True)
        
        # Save the updated DataFrame to the Excel file
        with pd.ExcelWriter(excel_filename, engine='openpyxl', mode='w') as writer:
            updated_df.to_excel(writer, index=False)
        
        # Style the Excel file
        style_excel(excel_filename)

    except FileNotFoundError:
        # If the file does not exist, create a new Excel file with slno starting from 1
        new_data = {
            'SLNO': [1],
            'DATE': [datetime.now().strftime('%Y-%m-%d')],
            'BATCHFOLDER': [batchfolder],
            'FILENAME': [filename],
            'STATUS': [status],
            'REMARK': [remark]
        }
        new_df = pd.DataFrame(new_data)
        new_df.to_excel(excel_filename, index=False, engine='openpyxl')
        
        # Style the Excel file
        style_excel(excel_filename)

# Example of appending logs
append_log('file1.txt', 'Success', 'Processed successfully', 'Batch1')
append_log('file2.txt', 'Failed', 'File not found', 'Batch2')
```

```python
# apply clustering 
cluster_labels = cluster_obj.predict(preprocossed_stage_one_data)
len(cluster_labels)
```

```python
# Concatenate the cluster labels with the preprocessed data
preprocossed_stage_one_data = pd.concat(
    [preprocossed_stage_one_data, pd.Series(cluster_labels, name='Cluster')], 
    axis=1
)

# Display the updated DataFrame
preprocossed_stage_one_data
```

```python
# combile wafer names final file
preprocossed_stage_one_data = pd.concat(
    [preprocossed_stage_one_data, preprocessing_stage_one_data_with_wafer_column[['Wafer']]],
    axis=1
)

# Now you can copy the DataFrame if needed
final_cluster_data = preprocossed_stage_one_data.copy()

# Display the final data
final_cluster_data
```

```python
# Perform predictions cluster-wise and combine all the predictions
prediction_files = []
for cluster_no in range(final_cluster_data['Cluster'].nunique()):
    cluster_sub_folder_path = Path(os.path.join(prediction_pipeline_config.best_models_path, f'Cluster_{cluster_no}'))
    print(f"Getting the data from Cluster_{cluster_no}")
    
    # Extract the data for this cluster
    cluster_data = final_cluster_data[final_cluster_data['Cluster'] == cluster_no]
    prediction_data = cluster_data.drop(columns=['Wafer', 'Cluster'])
    
    # Load objects
    scalar_obj_path = cluster_sub_folder_path / prediction_pipeline_config.standard_scalar_obj_name
    pca_obj_path = cluster_sub_folder_path / prediction_pipeline_config.pca_obj_name
    model_path = cluster_sub_folder_path / prediction_pipeline_config.best_model_name
    scalar_obj = load_obj(scalar_obj_path)
    pca_obj = load_obj(pca_obj_path)
    model_obj = load_obj(model_path)
    
    # Transform data
    scaled_data = scalar_obj.transform(prediction_data)
    pca_data = pca_obj.transform(scaled_data)
    predictions = model_obj.predict(pca_data)
    
    # Ensure index alignment by resetting the index
    cluster_wafer = cluster_data['Wafer'].reset_index(drop=True)
    predictions_series = pd.Series(predictions, name='Output').reset_index(drop=True)
    
    # Concatenate predictions with wafer data
    final_prediction = pd.concat([cluster_wafer, predictions_series], axis=1)
    
    print(f"Cluster {cluster_no} - Number of Wafer IDs: {len(cluster_wafer)}")
    print(f"Cluster {cluster_no} - Number of Predictions: {len(predictions_series)}")
    
    # Store the final prediction for this cluster
    prediction_files.append(final_prediction)

# Combine all the predictions from different clusters
final_predictions_combined = pd.concat(prediction_files, axis=0).reset_index(drop=True)

# Display the final combined predictions
final_predictions_combined
```
